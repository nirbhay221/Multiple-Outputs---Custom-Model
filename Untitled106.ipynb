```
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input,Dense
from sklearn.model_selection import train_test_split
```

```
def format_output(data):
  y1= data.pop('Y1')
  y1 =np.array(y1)
  y2 = data.pop('Y2')
  y2 = np.array(y2)
  return y1,y2
```

```
def norm(x):
  return (x-train_stats['mean'])/train_stats['std']
```

```
def plot_diff(y_pred,y_true,title=''):
  plt.scatter(y_true,y_pred)
  plt.title(title)
  plt.xlabel('True Values')
  plt.ylabel('Predictions')
  plt.axis('equal')
  plt.axis('square')
  plt.xlim(plt.xlim())
  plt.ylim(plt.ylim())
  plt.plot([-100,100],[-100,100])
  plt.show()
```

```
def plot_metrics(metrics_name,title,ylim=5):
  plt.title(title)
  plt.ylim(0,ylim)
  plt.plot(history.history[metrics_name],color='blue',label=metrics_name)
  plt.plot(history.history['val_'+metrics_name],color ='green',label ='val_'+metrics_name)
  plt.show()
```

```

URL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00242/ENB2012_data.xlsx'
```

```
df = pd.read_excel(URL)
df = df.sample(frac=1).reset_index(drop=True)
train,test = train_test_split(df,test_size=0.2)
train_stats = train.describe()
```

```
train_stats.pop('Y1')
train_stats.pop('Y2')
train_stats = train_stats.transpose()
train_y= format_output(train)
test_y = format_output(test)
norm_x_train = norm(train)
norm_x_test = norm(test)
```

```
input_layer = Input(shape=(len(train.columns),))
first_dense = Dense(128,activation='relu')(input_layer)
second_dense =Dense(128,activation='relu')(first_dense)
y1_output= Dense(1,name='y1_output')(second_dense)
third_dense =Dense(64,activation='relu')(second_dense)
y2_output=Dense(1,name='y2_output')(third_dense)
model= Model(inputs=input_layer,outputs= [y1_output,y2_output])
print(model.summary())
```

```
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.001),loss={'y1_output':'mse','y2_output':'mse'},metrics={
    'y1_output':tf.keras.metrics.RootMeanSquaredError(),
    'y2_output':tf.keras.metrics.RootMeanSquaredError()
})
```

```
history= model.fit(norm_x_train,train_y,epochs=500,batch_size=10,validation_data=(norm_x_test,test_y))
```
